```python
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(cps5s.wage,bins=30,edgecolor='w')
plt.xlabel("Wage")
plt.ylabel("Count");
```

```python
import seaborn as sns; sns.set_style('whitegrid',  {"grid.color": '.95'}) 

ax = sns.histplot(cps5s.wage,bins=30,stat='percent')
sns.rugplot(cps5s.wage,ax=ax)
ax.set_ylim(-1.5,31);
```

```python
np.random.seed(42)  # to give the same pseudo-random sequence each time

ts = pd.Series(np.random.randn(500),
               index=pd.date_range('1/1/2022', periods=500)).cumsum()

plt.plot(ts)
plt.xlabel("Date")
plt.xticks(rotation=30)
plt.ylabel("Value")
plt.show()
```

```python
from scipy.stats import norm

z = np.linspace(-4,4,200)
y = norm.pdf(z)
cdf = norm.cdf(z)

fig, ax = plt.subplots(2,1,sharex='col')

ax[0].plot(z,y)
ax[1].plot(z,cdf)

ax[1].set_xlabel("x")
ax[1].set_ylabel("cdf")
ax[0].set_ylabel("pdf");
```

```python
from scipy.stats import norm
print(norm.cdf(1.45))
```
